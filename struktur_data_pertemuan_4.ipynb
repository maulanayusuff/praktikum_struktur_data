```python
#queue dengan list
queue= [] #membuat variabel queue

queue.append("a") #menambahkan elemen a
queue.append("b") #menambahkan elemen b
queue.append("c") #menambahkan elemen c
queue.append("d") #menambahkan elemen d
queue.append("e") #menambahkan elemen e

print("Initial queue") #menampilkan "initial queue"
print(queue) #menampilkan queue

print("\nElements poped from queue")
print(queue.pop(0)) #menghapus queue pertama
print(queue.pop(0)) #menghapus queue pertama
print(queue.pop(0)) #menghapus queue pertama

print("\nqueue after elements are poped") #menampilkan "queue after elements are poped"
print(queue) #menampilkan queue
```

```python
#stack dengan collections.deque
from collections import deque #mengimpor deque dari collections

queue= deque() #membuat variabel queue

queue.append("a") #menambahkan elemen a
queue.append("b") #menambahkan elemen b
queue.append("c") #menambahkan elemen c
queue.append("d") #menambahkan elemen d
queue.append("e") #menambahkan elemen e

print("Initial queue") #menampilkan "initial queue"
print(queue) #menampilkan queue

print("\nElements poped from queue") #menampilkan "Elements poped from queue"
print(queue.popleft()) #menghapus queue pertama
print(queue.popleft()) #menghapus queue pertama
print(queue.popleft()) #menghapus queue pertama

print("\nqueue after elements are poped") #menampilkan "queue after elements are poped"
print(queue) #menampilkan queue
```

```python
#queue dengan queue.Queue
from queue import Queue #mengimpor Queue dari queue

qu= Queue(maxsize=5) #membuat variabel queue

print(qu.qsize()) #menampilkan jumlah item dalam queue saat ini

qu.put("a") #menambahkan elemen a
qu.put("b") #menambahkan elemen b
qu.put("c") #menambahkan elemen c
qu.put("d") #menambahkan elemen d
qu.put("e") #menambahkan elemen e

print("full:", qu.full()) #mengecek apakah jumlah item dalam queue full
print("size:", qu.qsize()) #menampilkan jumlah item dalam queue saat ini

print("\nElements poped from the queue") #menampilkan "Elements poped from the queue""
print(qu.get()) #menghapus queue pertama
print(qu.get()) #menghapus queue pertama
print(qu.get()) #menghapus queue pertama

print("\nempty:", qu.empty()) #mengecek apakah item dalam queue kosong
print("full:", qu.full()) #mengecek apakah jumlah item dalam queue full
```

```python
class Queue: #membuat kelas queue
  def __init__(self): #menginisiasi queue
    self.queue= [] #membuat variabel queue sebagai list kosong

  def addtoqu(self, dataval): #metode untuk menambahkan elemen ke queue
    if dataval not in self.queue: #memeriksa apakah dataval sudah ada di queue
      self.queue.insert(0, dataval) #jika tidak, masukan dataval di awal list
      return True #mengembalikan true jika berhasil ditambahkan
    return False #mengembalikan false jika dataval sudah ada

  def removefromqu(self): #metode untuk menghapus elemen dari queue
    if len(self.queue) > 0: #memeriksa apakah queue tidak kosong
      return self.queue.pop() #menghapus elemen terakhir dari list
    return ("No elements in Queue") #menampilkan "No elements in Queue" jika queue kosong

  def size(self): #metode untuk mendapatkan ukuran queue
    return len(self.queue) #menampilkan jumlah elemen dalam list

thequeue= Queue() #membuat variabel dari kelas Queue bernama "thequeue"
print(thequeue.addtoqu("monday")) #menambahkan "monday" dan menampilkan True
print(thequeue.addtoqu("tuesday")) #menambahkan "tuesday" dan menampilkan True
print(thequeue.addtoqu("wednesday")) #menambahkan "wednesday" dan menampilkan True
print(thequeue.addtoqu("thrusday")) #menambahkan "thrusday" dan menampilkan True
print(thequeue.size()) #menampilkan jumlah data pada queue saat ini
print(thequeue.removefromqu()) #menghapus monday dan menampilkan elemen dari queue
print(thequeue.removefromqu()) #menghapus tuesday dan menampilkan elemen dari queue
print(thequeue.size()) #menampilkan jumlah data pada queue saat ini
```
