```python
from IPython.display import clear_output
import time

print("Selamat Datang di Program Menghitung BMI")

databmi= {}

def BMI():
  while True:
      try:
        time.sleep(0.5)
        nama= input("Masukan nama anda: ")
        bb= float(input("Masukan berat badan(kg): "))
        print("\bkg")
        tb= float(input("Masukan tinggi badan(cm): "))
        print("\bcm")

        nilaibmi= bb/(tb/100)**2

        print("Nilai BMI anda:", nilaibmi)

        if nilaibmi < 18.5:
          kategori= "kurus"
          print(f"\nBerat badan anda kurang, anda termasuk kategori {kategori}")
          print("\nTubuhmu membutuhkan lebih banyak energi,")
          print("cobalah menambah asupan makanan bergizi seperti protein, karbohidrat kompleks, dan lemak sehat.")
          print("Makanlah secara teratur dan tambahkan camilan bernutrisi di antara jam makan.")
          print("Ingat, proses menaikkan berat badan yang sehat butuh waktu.")
          print("Pelan-pelan aja, setiap langkah kecil tetap membuatmu lebih kuat!")
        elif 18.5 <= nilaibmi <= 24.9:
          kategori= "normal"
          print(f"\nBerat badan anda ideal, anda termasuk kategori {kategori}")
          print("\nBerat badanmu berada dalam kisaran yang sehat.")
          print("Pertahankan pola makan seimbang dan tetap aktif berolahraga secara rutin.")
          print("Jaga tidur yang cukup dan kelola stres agar tubuh tetap prima.")
          print("Kamu sudah berada di jalur yang tepat, lanjutkan kebiasaan bagusmu!")
        elif 25.0 <= nilaibmi <= 29.9:
          kategori= "gemuk"
          print(f"\nBerat badan anda berlebih, anda termasuk kategori {kategori}")
          print("\nTubuhmu memberikan sinyal untuk mulai menata ulang pola makan dan aktivitas harian.")
          print("Kurangi konsumsi makanan tinggi gula dan gorengan, lalu perbanyak sayur, buah, dan air putih.")
          print("Mulai saja dari aktivitas ringan seperti berjalan kaki, yang penting konsisten.")
          print("Ingat, nggak harus sempurna, yang penting terus maju sedikit demi sedikit!")
        else:
          kategori= "obesitas"
          print(f"\nBerat badan anda terlalu berlebih, anda termasuk kategori {kategori}")
          print("\nSekarang adalah waktu yang baik untuk mulai melakukan perubahan kecil yang berdampak besar.")
          print("Kendalikan porsi makan, pilih makanan yang lebih sehat, dan coba bergerak lebih aktif, misalnya dengan jalan kaki rutin.")
          print("Jika merasa membutuhkan bantuan, tidak apa-apa mencari dukungan dari orang terdekat atau profesional.")
          print("Perubahan besar selalu dimulai dari langkah kecil, kamu pasti mampu!")

        databmi[nama]= {
            "bb": bb,
            "tb": tb,
            "BMI": nilaibmi,
            "kategori": kategori
        }

      except:
        print("Input tidak valid, pastikan bb dan tb adalah angka")
        continue

      while True:
        lagi= input("\nApakah anda ingin menghitung BMI lagi (y/n)?: ").lower()
        if lagi == "y":
          clear_output(wait= True)
          print("Selamat Datang di Program Menghitung BMI")
          break
        elif lagi == "n":
          clear_output(wait= True)
          print("Data BMI tercatat\n")
          print("-"*68)
          print(f"{"Nama":<15}{"Berat(kg)":<15}{"Tinggi(cm)":<15}{"BMI":<15}{"Kategori":<8}")
          print("-"*68)
          for nama, data in databmi.items():
            print(f"{nama:<15}{data["bb"]:<15}{data["tb"]:<15}{data["BMI"]:<15.2f}{data["kategori"]:<8}")
          print("-"*68)
          print("\nProgram selesai, terima kasih")
          return
        else:
          print("\nInput tidak valid, pastikan pilihan adalah y/n")

BMI()
```
