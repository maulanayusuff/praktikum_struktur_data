```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None
        self.prevval = None

class DoubleLinkedList:
    def __init__(self):
        self.headval = None

    def listprint(self):
        if self.headval is None:
            print("Double Linked List kosong")
            return

        current = self.headval
        while current:
            print(current.dataval)
            current = current.nextval

    def addinend(self, newdata):
        new_node = Node(newdata)
        if self.headval is None:
            self.headval = new_node
            return

        current = self.headval
        while current.nextval:
            current = current.nextval
        current.nextval = new_node
        new_node.prevval = current

    def addafter(self, target_data, new_data):
        if self.headval is None:
            print(f"List kosong, tidak bisa menyisipkan {new_data} setelah {target_data}")
            return

        current = self.headval
        while current:
            if current.dataval == target_data:
                new_node = Node(new_data)
                new_node.nextval = current.nextval
                new_node.prevval = current
                if current.nextval:
                    current.nextval.prevval = new_node
                current.nextval = new_node
                print(f"Berhasil menyisipkan {new_data} setelah {target_data}")
                return
            current = current.nextval
        print(f"Node dengan data '{target_data}' tidak ditemukan.")

    def deleteend(self):
        if self.headval is None:
            print("Double Linked List kosong!")
            return

        if self.headval.nextval is None:
            self.headval = None
            return

        current = self.headval
        while current.nextval.nextval:
            current = current.nextval
        current.nextval = None

    def delete_by_value(self, data_to_delete):
        if self.headval is None:
            print("Double Linked List kosong, tidak ada yang bisa dihapus!")
            return

        current = self.headval

        # Case 1: Head node is the one to be deleted
        if current.dataval == data_to_delete:
            self.headval = current.nextval
            if self.headval:
                self.headval.prevval = None
            print(f"Node '{data_to_delete}' berhasil dihapus.")
            return

        # Case 2: Node to be deleted is not the head
        while current and current.dataval != data_to_delete:
            current = current.nextval

        if current is None: # Data not found
            print(f"Node dengan data '{data_to_delete}' tidak ditemukan.")
            return

        # Bypass the current node (node to be deleted)
        if current.prevval:
            current.prevval.nextval = current.nextval
        if current.nextval:
            current.nextval.prevval = current.prevval
        print(f"Node '{data_to_delete}' berhasil dihapus.")


# ==== Contoh Penggunaan Double Linked List ====

dll = DoubleLinkedList()

dll.addinend("senin")
dll.addinend("rabu")
dll.addinend("kamis")

print("Data awal (Double Linked List):")
dll.listprint()

print("\n==============================")
print("Data setelah menyisipkan 'selasa' setelah 'senin':")
dll.addafter("senin", "selasa")
dll.listprint()

print("\n==============================")
print("Data setelah menghapus 'rabu':")
dll.delete_by_value("rabu")
dll.listprint()
```
