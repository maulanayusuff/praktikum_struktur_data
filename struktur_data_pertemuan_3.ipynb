```python
#stack dengan list
stack= [] #membuat variabel stack

stack.append("a") #menambahkan elemen a
stack.append("b") #menambahkan elemen b
stack.append("c") #menambahkan elemen c
stack.append("d") #menambahkan elemen d
stack.append("e") #menambahkan elemen e

print("Initial stack") #menampilkan "initial stack"
print(stack) #menampilkan stack

print("\nElements poped from stack")
print(stack.pop()) #menghapus stack terakhir
print(stack.pop()) #menghapus stack terakhir
print(stack.pop()) #menghapus stack terakhir

print("\nStack after elements are poped") #menampilkan "Stack after elements are poped"
print(stack) #menampilkan stack
```

```python
#stack dengan collections.deque
from collections import deque #mengimpor deque dari collections

stack= deque() #membuat variabel stack

stack.append("a") #menambahkan elemen a
stack.append("b") #menambahkan elemen b
stack.append("c") #menambahkan elemen c
stack.append("d") #menambahkan elemen d
stack.append("e") #menambahkan elemen e

print("Initial stack") #menampilkan "initial stack"
print(stack) #menampilkan stack

print("\nElements poped from stack") #menampilkan "Elements poped from stack"
print(stack.pop()) #menghapus stack terakhir
print(stack.pop()) #menghapus stack terakhir
print(stack.pop()) #menghapus stack terakhir

print("\nStack after elements are poped") #menampilkan "Stack after elements are poped"
print(stack) #menampilkan stack
```

```python
#stack dengan queue.LifoQueue
from queue import LifoQueue #mengimpor LifoQueue dari queue

stack= LifoQueue(maxsize=5) #membuat variabel stack

print(stack.qsize()) #menampilkan jumlah item dalam stack saat ini

stack.put("a") #menambahkan elemen a
stack.put("b") #menambahkan elemen b
stack.put("c") #menambahkan elemen c
stack.put("d") #menambahkan elemen d
stack.put("e") #menambahkan elemen e

print("full", stack.full()) #mengecek apakah jumlah item dalam stack full
print("size", stack.qsize()) #menampilkan jumlah item dalam stack saat ini

print("\nElements poped from stack") #menampilkan "Elements poped from stack""
print(stack.get()) #menghapus stack terakhir
print(stack.get()) #menghapus stack terakhir
print(stack.get()) #menghapus stack terakhir

print("\nempty:", stack.empty()) #mengecek apakah item dalam stack kosong
```

```python
#stack dengan single linked list
class node: #membuat kelas node
  def __init__(self, value): #menginisiasi value stack
    self.value= value #menyimpan value stack
    self.next= None #tidak memberi node selanjutnya

class stack: #membuat kelas stack
  def __init__(self):
    self.head= node("head") #membuat node pertama
    self.size= 0

  def  __str__(self): #mendefinisikan node
    cur= self.head.next
    elements = []
    while cur:
      elements.append(str(cur.value))
      cur= cur.next
    return "->".join(elements)

  def getsize(self): #mengecek size stack
    return self.size

  def isempty(self): #mengecek apakah stack kosong
    return self.size == 0

  def peek(self): #melihat stack teratas atau terakhir
    if self.isempty():
      raise Exception("peeking from an empty stack")
    return self.head.next.value

  def push(self, value): #mengisi stack
    newnode= node(value)
    newnode.next= self.head.next
    self.head.next= newnode
    self.size += 1

  def poop(self): #mengambil isi stack
    if self.isempty():
      raise Exception("pooping from an empty stack")
    remove= self.head.next
    self.head.next= self.head.next.next
    self.size -= 1
    return remove.value

if __name__ == "__main__":
  stack_instance= stack() #membuat variabel stack
  for i in range(1, 11): #menambahkan angka 1 sampai 10 ke dalam stack
    stack_instance.push(i)
  print(f"stack: {stack_instance}") #menampilkan isi stack saat ini

  for _ in range(1, 6): #untuk 5 angka terakhir
    removed_value= stack_instance.poop() #hapus 5 angka terakhir dari stack
    print(f"Pop: {removed_value}") #menampilkan angka yang dihapus
  print(f"stack: {stack_instance}") #menampilkan isi stack saat ini
```
